# Corona Virus data

```python
import seaborn as sns
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

ill = pd.read_csv('resources/time_series_covid19_confirmed_global.csv')
toll = pd.read_csv('resources/time_series_covid19_deaths_global.csv')
```

```python
# isr = ill[ill['Country/Region'] == 'Israel'].iloc[:,43:].iloc[0]
spain = ill[ill['Country/Region'] == 'Spain'].iloc[:,43:].iloc[0]
iran = ill[ill['Country/Region'] == 'Iran'].iloc[:,43:].iloc[0]
italy = ill[ill['Country/Region'] == 'Italy'].iloc[:,43:].iloc[0]
france = ill[ill['Country/Region'] == 'France'].iloc[:,43:].sum()

plt.figure(figsize=(20,10))

# plotting the points  
# plt.plot(isr, label='ISR')
plt.plot(spain, label='SPA', color='yellow')
plt.plot(iran, label='IRA', color='green')
plt.plot(italy, label='ITA', color='red')
plt.plot(france, label='FRA', color='blue')
plt.grid()

# naming the x axis 
plt.xlabel('ךיראת')
# naming the y axis 
plt.ylabel('םילוחה רפסמ')
  
# giving a title to my graph 
plt.title(' הנורוקה תוטשפתה') 
plt.legend()
# function to show the plot 
plt.show() 
```
